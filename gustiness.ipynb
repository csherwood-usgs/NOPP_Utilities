```python
import numpy as np
import matplotlib.pyplot as plt

def pcoord(x, y):
    """
    Convert x, y to polar coordinates r, az (geographic convention)
    r,az = pcoord(x, y)
    """
    r = np.sqrt(x**2 + y**2)
    az = np.degrees(np.arctan2(x, y))
    # az[where(az<0.)[0]] += 360.
    az = (az+360.)%360.
    return r, az


def xycoord(r, az):
    """
    Convert r, az [degrees, geographic convention] to rectangular coordinates
    x,y = xycoord(r, az)
    """
    x = r * np.sin(np.radians(az))
    y = r * np.cos(np.radians(az))
    return x, y
```

```python
t = np.arange(0, 1801, 1)
ubar = 5
vbar = 5
ua = 1
va = 1
up = 60
vp = 60
```

```python
u = ubar + ua * np.cos(t*2.*np.pi/up)
v = vbar + va * np.cos(t*2.*np.pi/vp)
#v = vbar*np.ones_like(u)
U = np.sqrt(u**2 + v**2)
stdu = np.std(u)
stdv = np.std(v)
stdU = np.std(U)
Gnum = np.sqrt(stdu**2 + stdv**2)
Gdenom = np.mean(U)
print('Std u, std v, std U, Ubar, G:', stdu, stdv, stdU, np.mean(U), stdU/np.mean(U))
print('Gnum, Gdenom, Gnum/Gdenom:',Gnum, Gdenom, Gnum/Gdenom)
```

```python
spd, dirw = pcoord(u,v)
fig, ax = plt.subplots(2,1)
ax[0].plot(U)
ax[1].plot(dirw)
```
